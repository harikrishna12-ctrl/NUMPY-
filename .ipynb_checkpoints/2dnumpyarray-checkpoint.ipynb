```python
import numpy as np
```

```python
arr2d= np.array([[1,2,3],[2,4,5],[3,4,5]])
print(arr2d,arr2d.shape)
```

```python

arr1=np.zeros((2,4))
print(arr1,arr1.shape)
```

```python
arr1=np.ones((2,4))
print(arr1,arr1.shape)
```

```python
arr1=np.full((2,4),1000)
print(arr1,arr1.shape)
```

```python
arr1=np.eye((3))
print(arr1,arr1.shape) #identity matrix
```

```python
arr1=np.arange(0,11,2)
print(arr1,arr1.shape)
```

```python
arr1=np.linspace(1,100,4)
print(arr1,arr1.shape)  # iwant a number between 1 to 100 but i need only 4 numbers  
```

```python
#properties of numpy
arr= np.array([[1,2,3],[2,3,4],[2,3,4]])
print(arr.shape)
print(arr.size)
print(arr.dtype)
print(arr.ndim)
float_arr= arr.astype(np.float64)
print(float_arr,float_arr.dtype)
```
