```python
import numpy as np
```

```python
num = np.arange(1,11)
print(num)
```

```python
num = np.zeros(10)
print(num)
```

```python
num = np.ones(10)
print(num)
```

```python
num = np.linspace(0,1,10)
print(num)
```

```python
num = np.full((3,3),7)
print(num)
```

```python
num = np.array([10,20,30,40,50])
print(num[2])

#last 2 elemetns
print(num[-1])
print(num[-2])
```

```python
num = np.array([10,20,30,40,50])
print(num[ : :2])
```

```python
#reversing a array
num = np.array([10,20,30,40,50])
print(num[len(num)::-1])
```
